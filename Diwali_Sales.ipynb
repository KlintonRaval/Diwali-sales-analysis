```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read CSV file
df = pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
df
```

```python
# info is used to describe the columns, not-null value, DataType, memory usage, etc.
df.info()
```

```python
# # describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
df.describe()
```

```python
#drop unwanted columns
df.drop(['Status', 'unnamed1'], axis=1, inplace = True)
df
```

```python
df.columns
```

```python
# drop null values from the dataframe
df.dropna(inplace = True)
```

```python
# changing the datatype of Amount column from Float to integer
df['Amount'] = df['Amount'].astype('int')
```

```python
# to show datatype
df['Amount'].dtype
```

```python
# to rename the columns
df.rename(columns = {'Marital_Status':'Shadai'})
df
```

# Exploratary Data Analysis

### Gender Wise

```python
# Visualize Gender column by bar chart and also their counts.

sns.set(rc={'figure.figsize':(35,15)}, font_scale = 3.5)
gender = sns.countplot( x = 'Gender', data = df)
for bars in gender.containers:
    gender.bar_label(bars)
```

```python
# filtering by Groupby for gender and amount

sales_gender  = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)
sales_gender
```

```python
# plotting bar chart for gender vs total amounts
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gender);
```

```python
gender = df.Gender
palette_color = sns.color_palette('dark')
plt.pie(gender, colors=palette_color)
```

***From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men***

### Age

```python
age = sns.countplot(x = 'Age Group', data = df, hue = 'Gender')
```

```python
# group by Total amount and Age
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sales_age
```

```python
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age);
plt.xticks(rotation = 90);
```

***From above graphs we can see that most of the buyers are of age group between 26-35 yrs female***

###  States

```python
# groupby State and Orders
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending= False)
sales_state
```

```python
sns.set(rc={'figure.figsize':(25,8)},font_scale = 2.5);
sns.barplot(data = sales_state, x = 'State', y = 'Orders');
plt.xticks(rotation=90);
```

### Marital Status
